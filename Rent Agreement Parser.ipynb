```python
import docx2txt
import spacy
from spacy import displacy
nlp = spacy.load('en_core_web_sm')
```

```python
dict = {}
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            dict[ent.text] = ent.label_
            print(ent.text + ' - ' + ent.label_ )
    else:
        print('No entities found')
```

```python
readText = docx2txt.process("./Validation_Data/Validation_Data/63793679-Rental-Agreement.pdf.docx")
doc=nlp(readText)
show_ents(doc)
```

```python
nlp.pipe_names
```

```python
####### Adding custom Names to pretrained Vocab
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
phrase_list = ["Sri Vishal Bhardwaj","Charnel Singh","Mr. Veerabrahmam Bathini","Prof. K. Parthasarathy","T.S.Krishna Iyengar ","Mr.Lingaiah Bathini "]
phrase_patterns = [nlp(text) for text in phrase_list]
matcher.add('newperson',None,*phrase_patterns)
found_matches = matcher(doc)
#print(found_matches)
from spacy.tokens import Span
PERSON = doc.vocab.strings[u"PERSON"]
found_matches
new_ents = [Span(doc, match[1], match[2], label = PERSON) for match in found_matches]
doc.ents = list(doc.ents) + new_ents
```

```python
####### Adding custom Names to pretrained Vocab
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
phrase_list = ["Rs.65,000/-","Rs.9,000/- "]
phrase_patterns = [nlp(text) for text in phrase_list]
matcher.add('newquantity',None,*phrase_patterns)
found_matches = matcher(doc)
from spacy.tokens import Span
QUANTITY = doc.vocab.strings[u"QUANTITY"]
found_matches
new_ents = [Span(doc, match[1], match[2], label = QUANTITY) for match in found_matches]
doc.ents = list(doc.ents) + new_ents
```

```python

displacy.render(doc, style='ent',jupyter='True')
```

```python
show_ents(doc)
print('\n','Names')
for d in dict:
    if dict[d] == 'PERSON':
        print(d,'\t',dict[d])
print('\n','Dates')
for d in dict:
    if dict[d] == 'DATE' and len(d)>10:
        print(d,'\t',dict[d])
print('\n','Agreement Value')
for d in dict:
    if dict[d] == 'QUANTITY':
        print(d,'\t',dict[d])
```
